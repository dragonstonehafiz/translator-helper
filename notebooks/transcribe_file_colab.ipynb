```python
import pathlib
import torch
import pysubs2
import whisper
```

```python
print("CUDA available:", torch.cuda.is_available())
print("Device:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "CPU")
```

```python
def transcribe_file(model, filepath: str, language="ja"):
    wav = pathlib.Path(filepath).with_suffix(".wav")
    print("Beginning Transcription")
    result = model.transcribe(str(wav), language=language)
    print("Transcription Complete")
    
    subs = pysubs2.SSAFile()
    style = subs.styles["Default"]
    style.fontsize = 46
    style.fontname = "Arial"
    subs.styles["Default"] = style

    for seg in result["segments"]:
        subs.events.append(pysubs2.SSAEvent(
            start=seg["start"] * 1000,
            end=seg["end"] * 1000,
            text=seg["text"]
        ))
        print(f"{seg['start']:.2f} - {seg['end']:.2f}: {seg['text']}")

    out_path = wav.with_suffix(".ass")
    subs.save(str(out_path), format_="ass")
```

```python
model = whisper.load_model("turbo")  # or "large" for more accuracy
```

```python
transcribe_file(model, "/workspace/02 ナイショの撮影会.wav")
```

```python
transcribe_file(model, "/workspace/03 責任とらせてもらいます.wav")
```
